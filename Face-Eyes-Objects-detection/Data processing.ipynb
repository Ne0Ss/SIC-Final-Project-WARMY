```python
images=[]
target_folder='DATA/dataset_yawn_full-image/test/no_yawn/'
for i in range(1,2622):
    img = cv2.imread(target_folder+str(i)+'.jpg')
    images.append(img)
images = [x for x in images if str(x) != 'None']
len(images)
```

```python
new_folder='DATA/dataset_yawn_processed/train/no_yawn/'

face_cascPath = 'haarcascade_frontalface_alt.xml'  #face detect model
faceCascade = cv2.CascadeClassifier(face_cascPath)

count=1
not_detected=[]
for img in images:
    frame = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Detect faces in the image
    faces = faceCascade.detectMultiScale(
        frame,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(90, 90),
    )
    if len(faces) >0:
        frame_tmp = img[faces[0][1]:faces[0][1] + faces[0][3], faces[0][0]:faces[0][0] + faces[0][2]:1, :]
        frame_tmp = cv2.resize(frame_tmp, (100, 100), interpolation=cv2.INTER_LINEAR)
        cv2.imwrite(new_folder+str(count)+'.jpg',frame_tmp)
    else:
        not_detected.append(count)
    count+=1
print(len(not_detected))
```

```python
new_folder='DATA/dataset_yawn_processed/test/no_yawn/'

count=1
not_detected=[]
for img in images:
    face_cascPath = 'haarcascade_frontalface_alt.xml'  #face detect model
    faceCascade = cv2.CascadeClassifier(face_cascPath)
    frame = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Detect faces in the image
    faces = faceCascade.detectMultiScale(
        frame,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(90, 90),
    )
    if len(faces) >0:
        frame_tmp = img[faces[0][1]:faces[0][1] + faces[0][3], faces[0][0]:faces[0][0] + faces[0][2]:1, :]
        frame_tmp = cv2.resize(frame_tmp, (100, 100), interpolation=cv2.INTER_LINEAR)
        cv2.imwrite(new_folder+str(count)+'.jpg',frame_tmp)
    else:
        face_cascPath = 'haarcascade_frontalface_alt2.xml'  #face detect model
        faceCascade = cv2.CascadeClassifier(face_cascPath)
        frame = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(
            frame,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(90, 90),
        )
        if len(faces) >0:
            frame_tmp = img[faces[0][1]:faces[0][1] + faces[0][3], faces[0][0]:faces[0][0] + faces[0][2]:1, :]
            frame_tmp = cv2.resize(frame_tmp, (100, 100), interpolation=cv2.INTER_LINEAR)
            cv2.imwrite(new_folder+str(count)+'.jpg',frame_tmp)
        else:
            face_cascPath = 'haarcascade_frontalface_alt_tree.xml'  #face detect model
            faceCascade = cv2.CascadeClassifier(face_cascPath)
            frame = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = faceCascade.detectMultiScale(
                frame,
                scaleFactor=1.1,
                minNeighbors=5,
                minSize=(90, 90),
            )
            if len(faces) >0:
                frame_tmp = img[faces[0][1]:faces[0][1] + faces[0][3], faces[0][0]:faces[0][0] + faces[0][2]:1, :]
                frame_tmp = cv2.resize(frame_tmp, (100, 100), interpolation=cv2.INTER_LINEAR)
                cv2.imwrite(new_folder+str(count)+'.jpg',frame_tmp)
            else:
                face_cascPath = 'haarcascade_profileface.xml'  #face detect model
                faceCascade = cv2.CascadeClassifier(face_cascPath)
                frame = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                faces = faceCascade.detectMultiScale(
                    frame,
                    scaleFactor=1.1,
                    minNeighbors=5,
                    minSize=(90, 90),
                )
                if len(faces) >0:
                    frame_tmp = img[faces[0][1]:faces[0][1] + faces[0][3], faces[0][0]:faces[0][0] + faces[0][2]:1, :]
                    frame_tmp = cv2.resize(frame_tmp, (100, 100), interpolation=cv2.INTER_LINEAR)
                    cv2.imwrite(new_folder+str(count)+'.jpg',frame_tmp)
                else:
                    face_cascPath = 'haarcascade_frontalface_default.xml'  #face detect model
                    faceCascade = cv2.CascadeClassifier(face_cascPath)
                    frame = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                    faces = faceCascade.detectMultiScale(
                        frame,
                        scaleFactor=1.1,
                        minNeighbors=5,
                        minSize=(90, 90),
                    )
                    if len(faces) >0:
                        frame_tmp = img[faces[0][1]:faces[0][1] + faces[0][3], faces[0][0]:faces[0][0] + faces[0][2]:1, :]
                        frame_tmp = cv2.resize(frame_tmp, (100, 100), interpolation=cv2.INTER_LINEAR)
                        cv2.imwrite(new_folder+str(count)+'.jpg',frame_tmp)
                    else:
                        print(count)
                        not_detected.append(count)
    count+=1
print(len(not_detected))
```
